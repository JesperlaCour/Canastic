```python
import numpy as np
import pandas as pd
import cv2
import os
```

```python
from random import shuffle
from tqdm import tqdm
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread, imshow, subplots, show
%matplotlib inline
```

```python
image = imread('Images/Unsorted/Image_10.jpeg')
images = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
imshow(images[0])
show()
```

```python
# ImageDataGenerator til at ndre/lave flere versioner af samme billede
```

```python
data_generator = ImageDataGenerator(rotation_range=90, brightness_range=(0.5, 1.5), shear_range=15.0, zoom_range=[.5, .9])
data_generator.fit(images)
image_iterator = data_generator.flow(images)

plt.figure(figsize=(16,16))
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(image_iterator.next()[0].astype('int'))
plt.show()
```

```python
data=[]

for i, img in tqdm(enumerate(os.listdir('Images/Unsorted/'))):
    label=i
    
    img = cv2.imread('Images/Unsorted/'+img, cv2.IMREAD_GRAYSCALE)
    #img = cv2.resize(img, (180, 180))
    if img is not None:
        imgs = img.reshape((1, img.shape[0], img.shape[1], 1))
        data_generator = ImageDataGenerator(rotation_range=90, brightness_range=(0.5, 1.5), shear_range=15.0, zoom_range=[.5, .9])
        data_generator.fit(imgs)
        image_iterator = data_generator.flow(imgs)
    
        for x in range(750):
            img_transformed=image_iterator.next()[0].astype('int')/255
            data.append([img_transformed, label])

shuffle(data)

np.save('Images/data.npy', data)
```
